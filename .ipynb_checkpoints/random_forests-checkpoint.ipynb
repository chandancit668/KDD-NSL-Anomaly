```python
from random import seed, randrange
from math import sqrt
import pandas as pd
import numpy as np

data_folder = "C:/Users/chaur/Documents/Fraud and Anomaly Detection/Assignment 1/NSL-KDD/"

data_train = pd.read_csv(data_folder+'csv_result-KDDTrain+.csv')
data_test = pd.read_csv(data_folder+'csv_result-KDDTest+.csv')

data_train.columns = [x.strip("\'") for x in data_train.columns.values]
data_test.columns = [x.strip("\'") for x in data_test.columns.values]

df_train = data_train
df_test = data_test

for i in range(0, len(data_train.columns.values)):
    print(i, data_train.columns.values[i]) 

print('\n SHAPE:\n\t',data_train.shape)

del df_train['id']
del df_test['id']
```

```python
actual_features = df_test['class']
print("Test data class values:\n",actual_features.value_counts())

print("\nCategorical Values in data:")
categorical_cols = ['protocol_type', 'service', 'flag', 'class'] 
data_train[categorical_cols].head()
```

# Converting categorical values to numeric
using Label Encoder

```python
from sklearn.preprocessing import LabelEncoder

categorical_values = df_train[categorical_cols]

trainlist = ['T']* (len(categorical_values))

len(trainlist)
categorical_values['is_train'] = trainlist

categorical_values

test_categorical = df_test[categorical_cols]
testlist = ['F']* (len(test_categorical))


test_categorical['is_train'] = testlist
test_categorical


categorical_enc = pd.concat([categorical_values, test_categorical])

categorical_enc

df_categorical_values_enc= categorical_enc.apply(LabelEncoder().fit_transform)
print(df_categorical_values_enc.head())
categorical_enc.head()

# categorical_values[categorical_cols].append(test_categorical)


# test_categorical.assign(is_train = testlist)

# # categorical_values[categorical_cols].append(test_categorical)

# # categorical_values

```

```python
categorical_cols
dataset_train = df_train
dataset_test = df_test

df_t = df_categorical_values_enc

trainlist, testlist = df_t[df_t['is_train'] == 1], df_t[df_t['is_train'] == 0]
# trainlist

for col in categorical_cols:
    dataset_train[col] = trainlist[col]
    dataset_test[col] = testlist[col]

# del dataset_test['is_train']
# del dataset_train['is_train']


print("No. of training observations: ", dataset_train.shape)

dataset_train.head()

print("No. of test subjects: ", dataset_test.shape)

```

```python
features = df_train.columns
features = features.drop('class')

y = dataset_train['class']

features
```

```python
from sklearn.ensemble import RandomForestClassifier


max_depth = 10
min_samples_split=10
n_jobs=2
random_state=0

clf = RandomForestClassifier()
clf.fit(dataset_train[features], y)


```

```python
prediction = np.array(clf.predict(dataset_test[features]))
prediction
```

```python
prediction_matrix = clf.predict_proba(data_test[features])[:]
prediction_matrix
```

```python
actual_values = np.array(dataset_test['class'].tolist())
count = 0
for i in range(len(actual_values)):
    if prediction[i] == actual_values[i]:
        count+=1

print("Accuracy: %.2f" % (count/len(actual_values))*100), '%' )
```
